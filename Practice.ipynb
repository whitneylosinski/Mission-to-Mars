```python
# Import tools
from splinter import Browser
from bs4 import BeautifulSoup as soup
```

```python
# Set up the path
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Visit the Quotes to Scrape site
url = 'http://quotes.toscrape.com/'
browser.visit(url)
```

```python
# Parse the HTML
html = browser.html
html_soup = soup(html, 'html.parser')
```

```python
# Scrape the Title
title = html_soup.find('h2').text
title
```

```python
# Scrap the top ten tags
tag_box = html_soup.find('div', class_='tags-box')
# tag_box
tags = tag_box.find_all('a', class_='tag')

for tag in tags:
    word = tag.text
    print(word)
```

```python
# Visit the Books to Scrape site
url = 'http://books.toscrape.com/'
browser.visit(url)
```

```python
# Parse the HTML
html = browser.html
url_soup = soup(html, 'html.parser')
```

```python
# Scrape the Title
books_list = url_soup.find('ul', class_='nav nav-list').get_text
books_list
```

```python
# Scrap the book url list
books_list = url_soup.find('ul', class_='nav nav-list')
urls = books_list.find_all('a')
```

```python
for url in urls:
    print(url)
```
